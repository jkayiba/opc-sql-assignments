```sql
WITH sales_2015 AS (
    SELECT
        ord_date
    FROM
        orders
    WHERE
        EXTRACT(YEAR FROM ord_date) = 2015
)
SELECT
    '2015' AS year_sales,
    ROW_NUMBER() OVER (ORDER BY ord_date DESC) AS num_of_sales
FROM
    sales_2015;

```

```sql
WITH monthly_sales_2015 AS (
    SELECT
        EXTRACT(MONTH FROM ord_date) AS month,
        EXTRACT(YEAR FROM ord_date) AS year
    FROM
        orders
    WHERE
        EXTRACT(YEAR FROM ord_date) = 2015
)
SELECT
    CONCAT(year, '-', month) AS year_month_sales,
    ROW_NUMBER() OVER (PARTITION BY month ORDER BY month) AS num_of_sales
FROM
    monthly_sales_2015;
```

Display the number of customers acquired each month. Your output should be limited to two columns. You can name the columns as you see fit.</span>

```sql
WITH customer_acquisition_2018 AS (
    SELECT
        EXTRACT(MONTH FROM cus_join_date) AS month,
        EXTRACT(YEAR FROM cus_join_date) AS year,
        cus_num
    FROM
        customers
    WHERE
        EXTRACT(YEAR FROM cus_join_date) = 2018
)
SELECT
    CONCAT('2018-', month) AS acquisition_month,
    COUNT(cus_num) AS num_of_customers
FROM
    customer_acquisition_2018
GROUP BY
    month
ORDER BY
    month;

```

```sql
SELECT
    c.cus_id,
    c.cus_num,
    c.cus_last_name,
    o.ord_ship_date,
    c.cus_join_date,
    o.ord_id,
    (o.ord_ship_date - c.cus_join_date) AS days_between
FROM
    customers c
JOIN
    orders o ON c.cus_id = o.cus_id  -- Assuming a relationship between customer and order
WHERE
    EXTRACT(YEAR FROM o.ord_ship_date) = 2018;
```

**Part 4**: Shimano has reached out to OPC and has asked that OPC provide a listing of all customers that purchased a bike with a combination of Shimano's race components XTR due to a critical recall issuance. Write a query that outputs the following elements to a table that can be exported and provided to Shimano. The table should be stored in the public schema.

- Customer ID

- Customer First Name
- Customer State

```sql
-- Drop the existing table
DROP TABLE IF EXISTS public.shimano_recall_customers;

-- Recreate the table and insert data
CREATE TABLE public.shimano_recall_customers (
    cus_id INT,
    cus_last_name VARCHAR(255),
    cus_first_name VARCHAR(255),
    cus_state VARCHAR(255),
    cus_phone VARCHAR(255)
);

-- Insert the relevant customer data
INSERT INTO public.shimano_recall_customers (cus_id, cus_last_name, cus_first_name, cus_state, cus_phone)
SELECT
    c.cus_id,
    c.cus_last_name,
    c.cus_first_name,
    c.cus_state,
    c.cus_phone
FROM
    customers c
JOIN
    orders o ON c.cus_id = o.cus_id
JOIN
    products p ON o.prod_id = p.prod_id
WHERE
    p.prod_description ILIKE '%Shimano XTR%';






```

**Part 5:** Write a query that will aggregate the total of all bikes sold. Group the output by the bike manufacturer. Your output should contain the following elements:

- Bike Manufacturer
- Bike Name
- Bike Description
- Bike Country of Origin
- Total of all sales

```sql
SELECT 
    p.prod_manufacturer AS "Bike Manufacturer",
    p.prod_name AS "Bike Name",
    p.prod_description AS "Bike Description",
    p.country_origin AS "Bike Country of Origin",
    SUM(o.order_tot) AS "Total of All Sales"
FROM 
    orders o
JOIN 
    products p ON o.prod_id = p.prod_id
GROUP BY 
    p.prod_manufacturer, p.prod_name, p.prod_description, p.country_origin
ORDER BY 
    "Total of All Sales" DESC;




```

```sql
SELECT
    p.prod_name AS "Bike Name",
    p.prod_manufacturer AS "Manufacturer",
    p.prod_price AS "Product Price",
    o.order_tot AS "Sale Price",
    o.order_tot - p.prod_price AS "Price Difference",
    AVG(p.prod_price) OVER () AS "Average Bike Price",
    SUM(o.order_tot) OVER () AS "Total Sales"
FROM
    orders o
JOIN
    products p ON o.prod_id = p.prod_id
ORDER BY
    "Price Difference" DESC;
```
